```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g.map_diag(sns.displot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# Regression Plots Lecture

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'],
           scatter_kws = {'s': 100} #size
          )
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex', row = 'time', hue = 'sex',
          aspect= 0.6 #space between grid
          )
```
